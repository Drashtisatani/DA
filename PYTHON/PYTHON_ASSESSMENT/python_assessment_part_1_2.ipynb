1. Clean the PRODUCTLINE column

```python
import pandas as pd

# Load dataset
df = pd.read_csv(
    '/Users/drashtisatani/Desktop/DA_python_Assessment/sales_data_sample.csv',
    encoding='latin1'
)

# Remove whitespace and apply title case
df['PRODUCTLINE'] = df['PRODUCTLINE'].str.strip().str.title()

# Check results
df[['PRODUCTLINE']].head()
```

2. Clean the SALES column and convert to float

```python
# Remove non-numeric characters and convert to float

df['SALES'] = (
    df['SALES']
    .astype(str)
    .str.replace(r'[^0-9.]', '', regex=True)
    .astype(float)
)

print(df['SALES'].head())
```

3. Create a dictionary with total sales by product line

```python
sales_dict = {}

for index, row in df.iterrows():
    product = row['PRODUCTLINE']
    sale = row['SALES']

    if product in sales_dict:
        sales_dict[product] += sale
    else:
        sales_dict[product] = sale

print(sales_dict)
```

4. Categorize product lines as "High Revenue Product"

```python
for product, total in sales_dict.items():
    if total > 100000:
        print(product, "-> High Revenue Product")
    else:
        print(product, "-> Normal Revenue Product")
```
