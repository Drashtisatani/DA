Import Libraries

```python
import pandas as pd
import numpy as np
```

Create the Load Function

```python
def load_data(file_path):
    try:
        df = pd.read_csv(file_path, encoding='latin1')
        print("Dataset Loaded Successfully")
        return df

    except Exception as e:
        print("Error Loading File:", e)
        return None
```

Load the CSV File

```python
file_path = "/Users/drashtisatani/Desktop/DA_python_Assessment/sales_data_sample.csv"

df = load_data(file_path)
```

Check the Data

```python
df.head()
```

Clean the PRODUCTLINE Column

```python
df["PRODUCTLINE"] = (
    df["PRODUCTLINE"]
    .astype(str)
    .str.strip()
    .str.title()
)

print("PRODUCTLINE cleaned successfully")
```

Clean and Convert SALES Column

```python
df["SALES"] = (
    df["SALES"]
    .astype(str)
    .str.replace(r"[^0-9.]", "", regex=True)
)

df["SALES"] = pd.to_numeric(df["SALES"], errors="coerce")

print("SALES column converted to float")
```

Check Missing Values

```python
print("Missing Values Report")
print("-" * 30)

print(df.isnull().sum())
```

Check Data Types

```python
print(df.dtypes)
```

Create Revenue Summary

```python
summary = (
    df.groupby("PRODUCTLINE")["SALES"]
    .sum()
    .sort_values(ascending=False)
)

summary
```

Categorize Products

```python
print("PRODUCT PERFORMANCE")
print("-" * 40)

for product, sales in summary.items():

    if sales > 100000:
        print(f"{product} --> High Revenue Product (Total Sales = {sales:,.2f})")
    else:
        print(f"{product} --> Low Revenue Product (Total Sales = {sales:,.2f})")
```

Detect Outliers

```python
Q1 = df["SALES"].quantile(0.25)
Q3 = df["SALES"].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - (1.5 * IQR)
upper_bound = Q3 + (1.5 * IQR)

outliers = df[
    (df["SALES"] < lower_bound) |
    (df["SALES"] > upper_bound)
]

print("Number of Outliers:", len(outliers))
```

Final Automated Report

```python
print("=" * 60)
print("AUTOMATED SALES INTEGRITY REPORT")
print("=" * 60)

print("Total Records       :", len(df))
print("Product Lines       :", df["PRODUCTLINE"].nunique())
print("Outlier Records     :", len(outliers))

print("\nTop Product Lines")
print("-" * 40)

print(summary.head())

print("=" * 60)
```
